```python
import pandas as pd
```

```python
labs = pd.read_csv('../../data/labs.csv')
first_diag = pd.read_csv('../../data/first_diagnosis.csv')
```

```python
merged = labs.merge(first_diag[['Patient', 'EntryDate']], on='Patient', how='left', suffixes=('', '_diag'))
```

```python
merged['CKD'] = merged['EntryDate_diag'].notna()
```

```python
merged
```

```python
filtered_labs = merged[(merged['EntryDate'] < merged['EntryDate_diag']) | merged['EntryDate_diag'].isna()]
```

```python
filtered_labs[filtered_labs['CKD'] == True]
```

```python
filtered_labs.sample(5)
```

```python
filtered_labs = filtered_labs.drop(columns=['EntryDate_diag'])
```

```python
filtered_labs.sample(10)
```

```python
filtered_labs = filtered_labs.sort_values(by=['Patient', 'EntryDate'])
```

```python
filtered_labs.to_csv("../../data/filtered_labs_full.csv", index=False)
```
